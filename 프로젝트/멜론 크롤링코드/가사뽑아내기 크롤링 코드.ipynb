```python
from selenium import webdriver   # selenium 프레임 워크에서 webdriver 가져오기
url = 'https://www.melon.com/chart/index.htm' # 멜론차트 페이지   # 접속할 웹 사이트 주소 (네이버)
driver = webdriver.Chrome('C:/anaconda3/chromedriver.exe')  # 크롬 드라이버로 크롬 켜기
driver.get(url) 

# 드라이버가 해당 url 접속
```

```python
# 가사 검색 드라이버


# 차트파인더 클릭
driver.find_element_by_xpath('//*[@id="gnb_menu"]/ul[1]/li[1]/div/div/button/span').click()

# 연도차트 클릭
driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/h4[3]/a').click()
time.sleep(1)

# 연대선택  클릭  
driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[1]/div[1]/ul/li[1]/span/label').click()
time.sleep(1)

# 연도선택  클릭   ( 이것만 따로 계속 바꿔줘야되요)
driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[2]/div[1]/ul/li/span/label').click()
time.sleep(1)

# 장르선택 국내종합 클릭
driver.find_element_by_xpath('//*[@id="d_chart_search"]/div/div/div[5]/div[1]/ul/li[2]/span/label').click()
time.sleep(1)

# 검색버튼 클릭
driver.find_element_by_xpath('//*[@id="d_srch_form"]/div[2]/button/span/span').click()


```

```python
html = driver.page_source # 드라이버 현재 페이지의 html 정보 가져오기 
                            # cf) requests.get(url)
soup = bs(html, 'lxml')
```

```python
# 제목 가져오기
title_elem = soup.select('div.ellipsis.rank01 > span > strong')
# titles = []

for i in title_elem:
  titles.append(i.text)

len(titles)
```

```python
# 가수명 가져오기
singer = soup.select('div.ellipsis.rank02 > span')
# singers = []
for i in singer:
    singers.append(i.text)
    
len(singers)
```

```python
def getlyrics(keyword):
    url = "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=" + keyword + "+가사"
    headers = { "User-Agent":"Mozilla/5.0" }
    html = requests.get(url, headers = headers ).text
    soup = bs(html, "html.parser")
    type07 = soup.select('div.intro_box > p')
    lyrics = str(type07)
    lyrics=lyrics.replace('<p class="text no_ellipsis type_center _content_text" style="max-height: 50rem; -webkit-line-clamp: 21;">','"').replace('<br>','","').replace('</br>','","').replace('<br/>','","').replace('</p>','"').replace(',"","","",""','')
    lyrics_list.append(lyrics)
    lyrics_list
    return lyrics_list
```

```python
for i in titles:
    getlyrics(i)
```

```python
df = pd.DataFrame({'제목':titles,'가수':singers,'가사':lyrics_list})
```
