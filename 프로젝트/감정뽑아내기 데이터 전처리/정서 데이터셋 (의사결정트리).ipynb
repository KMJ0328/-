```python
from konlpy.tag import Okt
okt = Okt()

def tokenize(doc):
    # norm은 정규화, stem은 근어로 표시하기를 나타냄
    return ['/'.join(t) for t in okt.pos(doc, norm=True, stem=True)]

all_docs = [(tokenize(df_data['댓글'][i]), df_data['정서'][i]) for i in range(0, len(df_data))]
```

```python
all_docs
```

```python
tokens = [t for d in all_docs for t in d[0]]
```

```python
tokens
```

```python
import nltk
from pprint import pprint
text = nltk.Text(tokens, name='NMSC')

# 전체 토큰의 개수
print(len(text.tokens))

# 중복을 제외한 토큰의 개수
print(len(set(text.tokens)))            

# 출현 빈도가 높은 상위 토큰 10개
pprint(text.vocab().most_common(10))
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
%matplotlib inline

font_fname = 'c:/windows/fonts/gulim.ttc'
font_name = font_manager.FontProperties(fname=font_fname).get_name()
rc('font', family=font_name)

plt.figure(figsize=(20,10))
text.plot(50)
```

```python
selected_words = [f[0] for f in text.vocab().most_common(10000)]

def term_frequency(doc):
    return [doc.count(word) for word in selected_words]

all_docs_train = [term_frequency(d) for d, _ in all_docs]
all_docs_test = [c for _, c in all_docs]
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

dt_clf = DecisionTreeClassifier()
skfold = StratifiedKFold(n_splits=6)
# StratifiedKFold의 split시 첫 인자는 데이터, 두 번째 인자는 결과가 들어가야 함으로 drop함수 사용
for train_idx, test_idx in skfold.split(x_train, x_test):
    X_train = x_train[train_idx]
    y_train = x_test[train_idx]
    
    dt_clf.fit(X_train, y_train)

X_train, X_test, y_train, y_test = train_test_split(x_train, x_test, test_size=0.2, random_state=62)
pred = dt_clf.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = round(accuracy_score(y_test, pred), 4)
accuracy
```

```python
def predict_target_emotion(target):
    token = tokenize(target)
    tf = term_frequency(token)
    data = np.expand_dims(np.asarray(tf).astype('float32'), axis=0)
    emotion = float(dt_clf.predict(data))
    return emotion
    print(emotion)
```
